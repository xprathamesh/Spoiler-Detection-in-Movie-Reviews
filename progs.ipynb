```
def convert_idx_to_chr(idx):
  return chr(idx + ord('A'))
```

```
def print_matrix(matrix):
  for i in range(len(matrix)):
    print('{} -> {}'.format(convert_idx_to_chr(i), matrix[i]))
```

```
def print_cluster(cluster, centroids):
  for key, val in cluster.items():
    temp = [convert_idx_to_chr(i) for i in val]
    print('{} -> {}'.format(centroids[key], temp))
```

```
def calc_euclid_dist(x,y):
  return math.sqrt((y[0]-x[0])**2 +(y[1]-x[1])**2)
```

```
def calc_updated_distances(inputs, centroids, dist):  
  update = False
  for i in range(len(inputs)):
    for j in range(len(centroids)):
      d = calc_euclid_dist(inputs[i], centroids[j])

      if d != dist[i][j]:
        dist[i][j] = d
        update = True

  return dist, update
```

```
def new_clusters(distances):
  clusters = defaultdict(list)

  for i in range(len(distances)):
    c = distances[i].index(min(distances[i]))
    clusters[c].append(i)

  return clusters
```

```
def calculate_clusters(inputs, centroids, max_itr):
  dist = [[0] * len(centroids) for _ in range(len(inputs))]
  itr = 1

  while(itr <= max_itr):
    print("================================")
    print("round: {}".format(itr))

    # calculate distance matrix
    dist, update = calc_updated_distances(inputs, centroids, dist)

    # if no updates, convergence reached
    if not update: 
      print('Reached convergence')
      break

    # print the distance matrix
    print(centroids)
    print_matrix(dist)

    # calculate clusters
    clusters = new_clusters(dist)
    print_cluster(clusters, centroids)

    # update the centroids
    centroids = []
    for key, val in clusters.items():
      sum_x, sum_y, tot = 0, 0, 0

      for i in val:
        point = inputs[i]
        sum_x += point[0]
        sum_y += point[1]
        tot += 1


      # calculate the mean of each point in the cluster
      centroids.append((sum_x / tot, sum_y / tot))

    itr += 1
```

```
calculate_clusters(inputs, centroids, 5)
```

# 2) Linear regression - 2D

```
def calc_w1(arr):
  n = len(arr)
  
  t1 = n * (sum(x * y for x,y in points))
  t2 = (sum(x for x,y in points)) * (sum(y for x,y in points))
  t3 = n * (sum(x * x for x,y in points))
  t4 = (sum(x for x,y in points)) ** 2
  w1 = (t1 - t2) / (t3 - t4)

  return w1
```

```
def calc_w0(arr, w1):
  n = len(arr)
  
  t5 = sum(y - w1 * x for x,y in points)
  w0 = t5 / n

  return w0
```

```
points = [(-1,2),(-5,-3),(-2,0)]
# points = [(8,5),(0,-2),(1,-3),(-8,-4)]

w1 = calc_w1(points)
w0 = calc_w0(points, w1)

print("w1 = {}".format(w1))
print("w0 = {}".format(w0))
```

Calculate RMSE and SSE

```
x,y = [x for x,y in points], [y for x,y in points]
y_pred = [xi * w1 + w0 for xi in x]
n = len(y_pred)

sse = sum([(yp-yi) ** 2 for yp, yi in zip(y_pred, y)])
mean_er = sse / n
rmse = math.sqrt(mean_er)

print("sse = {}".format(sse))
print("rmse = {}".format(rmse))
```

# 3) Association Analysis

Parameters count

```
def find_count(item_sets_list, check_items):
  count = 0
  
  for item_set in item_sets_list:
    if check_items.issubset(item_set):
      count += 1

  return count
```

```
# item_sets = [('bread','milk'),('bread','diaper','beer','eggs'),('milk','diaper','beer','coke'),('bread','milk','diaper','beer'),('bread','milk','diaper','coke')]

# # input_lhs = {'diaper','milk'}
# # input_rhs = {'beer'}

# input_lhs = {'bread'}
# input_rhs = {'diaper','milk'}

# item_sets = [('a','d','e'),('a','b','c','e'),('a','b','d','e'),('a','c','d','e'),('b','c','e'),('b','d','e'),('c','d'),('a','b','c'),('a','d','e'),('a','b','e')]
item_sets = [('a','b','c'),('a','b','d'),('b','c','e'),('b','c','d'),('a','c'),('a','c','d'),('a','c','e'),('a','d','e'),('c','d','e'),('c','d')]

input_lhs = {'b','d'}
input_rhs = {'e'}

min_sup = 7

n = len(item_sets)
input_set = {*input_lhs, *input_rhs}
```

```
sup_cnt_tot = find_count(item_sets, input_set)
sup_tot = sup_cnt_tot / n
print("support: {0:0.3f}".format(sup_tot))

sup_cnt_lhs = find_count(item_sets, input_lhs)
sup_cnt_rhs = find_count(item_sets, input_rhs)

conf_tot = sup_cnt_tot / sup_cnt_lhs
print("confidence: {0:0.3f}".format(conf_tot))

sup_lhs = sup_cnt_lhs / n
sup_rhs = sup_cnt_rhs / n

# lift > 1 for X -> indicates Y more likely if X
lift = conf_tot / sup_rhs
print("lift: {0:0.3f}".format(lift))

interest = sup_tot / (sup_lhs * sup_rhs)
print("interest: {0:0.3f}".format(interest))
```
